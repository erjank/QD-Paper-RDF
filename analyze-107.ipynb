```python
from qd_rdf import get_dots, data_path, plot_rdf
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rdfs = []
R = []
```

```python
L=5
dots=get_dots(data_path+"107/5a.tif",compare=True,microns=L, size=2,spacing=2)
b=1/np.sqrt(len(dots)/(L*1000*1000*L))
print(b)
R.append(b)
```

```python
rdfs.append(plot_rdf(dots,L,plot=True)[1])
```

```python
rdfs = np.asarray(rdfs)
mean_count = np.mean(rdfs,axis=0)
sem =  stats.sem(rdfs, axis=0, ddof=1)
r,rdf = plot_rdf(dots,L,plot=False)
plt.errorbar(1000*r/np.mean(R), mean_count, yerr=sem)
#plt.xlabel("r (μm)")
plt.xlabel("r/<R>")
plt.ylabel("<g(r)>")
plt.savefig("M107.png")
np.savetxt("M107.csv",(1000*r/np.mean(R),mean_count,sem),delimiter=',')
plt.show()
```
